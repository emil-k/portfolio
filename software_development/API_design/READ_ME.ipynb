```python
import requests
import pandas as pd
```

# Goal

The goal of this project is to take a csv spreadsheet with marketing campaign data, and expose it as an API endpoint, allowing querying, filtering, grouping and sorting.

# Scenario 1

Display impressions and clicks, before the 1st of June 2017, grouped by channel and country, sorted by clicks, descending. 

```python
first = "http://127.0.0.1:8000/get_report/?columns=impressions,clicks&to_date=2017-06-01&groupby=channel,country&sortby=clicks,desc"
r = requests.get(first)
df = pd.DataFrame(r.json())
```

```python
df
```

# Scenario 2

Show installs in May of 2017 on iOS, grouped by date, sorted by date, ascending.

```python
second = "http://127.0.0.1:8000/get_report/?columns=installs&os=ios&groupby=date&sortby=date,asc&from_date=2017-04-30&to_date=2017-06-01"
r = requests.get(second)
df = pd.DataFrame(r.json())
```

```python
df
```

# Scenario 3

Display revenue on June 1, 2017 in US, grouped by OS, sorted by revenue, descending.

```python
third = "http://127.0.0.1:8000/get_report/?columns=revenue&country=US&groupby=os&sortby=revenue,desc&from_date=2017-05-31&to_date=2017-06-02"
r = requests.get(third)
df = pd.DataFrame(r.json())
```

```python
df
```

# Scenario 4

Display CPI and spend in Canada grouped by channel ordered by CPI, descending.

```python
fourth = "http://127.0.0.1:8000/get_report/?columns=spend,cpi&country=CA&groupby=channel&sortby=cpi,desc"
r = requests.get(fourth)
df = pd.DataFrame(r.json())
```
